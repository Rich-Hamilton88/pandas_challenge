## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
gender_male = Heroes_df.loc[Heroes_df["Gender"]=="Male"]
#gender_male.head()

gender_female = Heroes_df.loc[Heroes_df["Gender"]=="Female"]
#gender_female.head()

gender_other = Heroes_df.loc[Heroes_df["Gender"]=="Other / Non-Disclosed"]
#gender_other.head()

count_males = len(gender_male["SN"].unique())
count_females = len(gender_female["SN"].unique())
count_others = len(gender_other["SN"].unique())

total_count = len(Heroes_df["SN"].unique())
#print(count_others)

percent_male = round(count_males / total_count *100,2)
percent_female = round(count_females / total_count *100,2)
percent_other = round(count_others / total_count *100,2)

summary3_df = pd.DataFrame([
                            {"": "Male", "Total Count": count_males, "Percentage of Players": percent_male}
                            ,{"": "Female", "Total Count": count_females, "Percentage of Players": percent_female}
                            ,{"": "Others / NonDisclosed", "Total Count": count_others, "Percentage of Players": percent_other}
                            ])
summary3_df

summary3_df.style.format({"Percentage of Players":"{0:.2f} %"})

#del summary3_df.index.name
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
bins = [0, 9, 14, 19, 24, 29, 34, 39, 100]

ages = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"] 

Heroes_df["Age Group"] = pd.cut(Heroes_df["Age"], bins, labels=ages)
Heroes_df

age_group = Heroes_df.groupby("Age Group")

purchase_count = age_group["Price"].count()

age_avg_pp = round(age_group["Price"].mean(),2)
age_total_pp = age_group["Price"].sum()
age_avgtotal_pp = round(age_avg_pp / age_total_pp *100,2)

summary6_df = pd.DataFrame({"Purchase Count": purchase_count, "Average Purchase Price": age_avg_pp,"Total Purchase Value": age_total_pp, "Avg Total Purchase per Person": age_avgtotal_pp})                
summary6_df

summary6_df.style.format({"Average Purchase Price":"${:,.2f}",
                                  "Total Purchase Value":"${:,.2f}",
                                  "Avg Total Purchase per Person":"${:,.2f}"})
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
items = Heroes_df[["Item ID", "Item Name", "Price"]]

popular_items = items.groupby(["Item ID", "Item Name"])

purchase_count = popular_items["Price"].count()

popular_tpv = popular_items["Price"].sum()

item_price = popular_tpv/purchase_count


summary8_df = pd.DataFrame({"Purchase Count": purchase_count, "Item Price": item_price,"Total Purchase Value": popular_tpv})                
summary8_df

summary8_df.style.format({"Item Price":"${:,.2f}", "Total Purchase Value":"${:,.2f}"})

summary8_df.sort_values(by='Purchase Count', ascending=False)
```
